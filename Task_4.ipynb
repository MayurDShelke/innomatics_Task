```python
# Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("❌ Minimum recharge should be ₹50")
        elif amount not in valid_plans:
            print("❌ Invalid plan. Choose from", valid_plans)
        else:
            print("✅ Recharge successful")
            break

validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"{item}: ⚠️ Reorder Alert")
        else:
            print(f"{item}: ✅ Stock OK")

products = {
    "Rice": 10,
    "Wheat": 20,
    "Sugar": 12
}

check_inventory(products)
```

```python
# Problem 3: Student Result Processing System
def check_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 70, 40, 50]
result = check_result(marks)

print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry
def calculate_fare(km, peak):
    fare = 50 + (km * 12)

    if peak == "yes":
        fare *= 1.25

    return fare

while True:
    km = float(input("Enter distance (km): "))
    peak = input("Peak hour? (yes/no): ")

    total_fare = calculate_fare(km, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker
def check_attendance(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P"]
print(check_attendance(attendance))
```

```python
# Problem 6: Password Strength Checker
def check_password(password):
    if len(password) < 8:
        return "Weak Password"

    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(check_password(pwd))
```
