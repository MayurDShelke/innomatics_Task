```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_post(likes_list):
    total_likes = 0

    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like

    print("Total Likes:", total_likes)

    # Check viral condition
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example input
likes = [200, 300, 150, 400]

analyze_post(likes)
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker
def check_rainfall(rainfall_list, required_level):
    total = 0

    # Calculate total rainfall
    for rain in rainfall_list:
        total += rain

    # Calculate average
    average = total / len(rainfall_list)

    print("Average Rainfall:", int(average))

    # Check adequacy
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example input
rainfall_data = [60, 70, 80, 75, 75]
required = 70

check_rainfall(rainfall_data, required)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def check_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example input
users = ["user1", "user2", "user3", "user1"]

check_duplicates(users)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


# Example
prices = [1500, 2500, 1800, 3200]
premium_crops(prices)
```

```python
# Problem 7: System Monitoring – Application Health Checker
def check_system(errors):
    print("Error Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example
check_system(7)
```

```python
# problem 9: E-Learning – Student Attendance Eligibility System
def check_attendance(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example
attendance = ["P", "P", "A", "P", "P"]
check_attendance(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def calculate_bill(units):
    bill = 0

    # Billing logic
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example
bill, status = calculate_bill(250)

print("Total Bill:", bill)
print("Usage Level:", status)
```
